# Экспоненциальное распрпеделение с *λ* = 1

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Генерация выборки объёмом 1000 

```python
expon_rw  = sts.expon()
sample = expon_rw.rvs(1000)
```

## Построение графика теоретической плотности и гистограммы сгенерированной выборки

```python
plt.hist(sample, density=True, bins=20, label="sample hist")

# на том же графике построим теоретическую плотность распределения экспоненциального распределения:
x = np.linspace(0, 6, 100)
pdf = expon_rw.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.9)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

## Генерация 1000 выборок объёмом *n* = 5 и построение гистограммы их средних

## Нормальное распределение с нужными параметрами для аппроксимации выборочных средних

Среднее экспоненциального распределения вычисляется по формуле: Ex = $1/\lambda$;  
дисперсия: Dx = 1/$\lambda^2$.
Поскольку я выбрал значение $\lambda = 1$, то параметры нормального распределения: $\mu=1$ и $\sigma=sqrt(1/5)$. 

При различных *n*: $\sigma=sqrt(1/n)$ (в дальнейшем этот пункт я буду опускать) 

```python
norm_rv = sts.norm(1, (1/5) ** 0.5)
new_x = np.linspace(0, 2, 100)
norm_pdf = norm_rv.pdf(new_x)
```

```python
sample_means = np.array([np.mean(np.random.choice(sample, size=5)) for _ in range(1000)])
```

```python
plt.hist(sample_means, density=True, bins=20, label="sample means hist")
pdf = expon_rw.pdf(new_x)
plt.plot(new_x, norm_pdf, label='normal pdf', alpha=0.8)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot()
```

## Генерация 1000 выборок объёмом *n* = 10 и построение гистограммы их средних

```python
sample_means = np.array([np.mean(np.random.choice(sample, size=10)) for _ in range(1000)])
```

## Нормальное распределение с нужными параметрами для аппроксимации выборочных средних

```python
norm_rv = sts.norm(loc=1, scale=((1/10) ** 0.5))
new_x = np.linspace(0, 2, 100)
norm_pdf = norm_rv.pdf(new_x)
```

```python
plt.hist(sample_means, density=True, bins=20, label="sample means hist")
plt.plot(new_x, norm_pdf, label='normal pdf', alpha=0.8)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot()
```

## Генерация 1000 выборок объёмом *n* = 50 и построение гистограммы их средних

```python
sample_means = np.array([np.mean(np.random.choice(sample, size=50)) for _ in range(1000)])
```

## Нормальное распределение с нужными параметрами для аппроксимации выборочных средних

```python
norm_rv = sts.norm(loc=1, scale=(1/50) ** 0.5)
new_x = np.linspace(0, 2, 100)
norm_pdf = norm_rv.pdf(new_x)
```

```python
plt.hist(sample_means, density=True, bins=20, label="sample means hist")
plt.plot(new_x, norm_pdf, label='normal pdf', alpha=0.8)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot()
```

Из построенных графиков следует вывод, что при росте *n* (объёме выборок) точность аппроксимации выборочных средних нормальным распределением увеличивается, что согласуется с теорией.
Более того, "пик" нормального распределения становится всё более узким.
